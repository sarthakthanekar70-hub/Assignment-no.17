```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# --- Q1. Preparing Features ---

# 1. Load your cleaned Ford Car Dataset (replace 'ford.csv' with your actual filename)
df = pd.read_csv("/content/cardata.csv")

# 2. Separate into Independent Features (X) and Dependent Feature (Y)
# 'price' is what we want to predict (Y), the rest are features (X)
X = df.drop(columns=["Selling_Price"])
Y = df["Selling_Price"]

# 3. Print the shape of X and Y to confirm separation
print("--- Q1: Shapes ---")
print("Shape of X (Features):", X.shape)
print("Shape of Y (Target):", Y.shape)
```

```python
# --- Q2. One-Hot Encoding ---

# 1. Identify all categorical columns in X
print("\n--- Q2: Categorical Columns ---")
categorical_cols = X.select_dtypes(include=["object", "category"]).columns
print("Categorical columns found:", list(categorical_cols))

# 2. Perform One-Hot Encoding using pd.get_dummies()
X_encoded = pd.get_dummies(X, columns=categorical_cols)

# 3. Convert the encoded data to integer type (True/False becomes 1/0)
X_encoded = X_encoded.astype(int)

# 4. Show the first 5 rows after encoding
print("\nFirst 5 rows of encoded data:")
print(X_encoded.head())
```

```python
# --- Q3. Feature Scaling ---

# 1. Apply Standard Scaling on the numerical columns
numerical_cols = ["Year", "Present_Price", "Kms_Driven", "Owner"]

# Create a copy of the encoded dataframe to safely scale values
X_scaled = X_encoded.copy()

# Initialize the StandardScaler
scaler = StandardScaler()

# 2. Use StandardScaler from sklearn.preprocessing
# We fit and transform only the numerical columns
X_scaled[numerical_cols] = scaler.fit_transform(X_scaled[numerical_cols])

# 3. Show the first 5 rows of the scaled features
print("\n--- Q3: Scaled Features (First 5 rows) ---")
print(X_scaled.head())
```

```python
# --- Q4. Train-Test Split ---

# 1. Split the preprocessed data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, Y, test_size=0.33, random_state=42
)

# 2. Print the shape of X_train, X_test, y_train, and y_test
print("\n--- Q4: Train-Test Split Shapes ---")
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# --- Q5. Building Linear Regression Model ---

# 1. Train a Linear Regression model using sklearn
model = LinearRegression()

# 2. Fit the model on the training data
model.fit(X_train, y_train)

# 3. Print the intercept and coefficients of the model
print("--- Q5: Model Parameters ---")
print("Intercept (Starting baseline price):", model.intercept_)
print("\nCoefficients (Weight of each feature):")
for col, coef in zip(X_train.columns, model.coef_):
    print(f"{col}: {coef}")
```

```python
# --- Q6. Making Predictions ---

# 1. Use the trained model to make predictions on the test data
# 2. Store predictions in y_pred
y_pred = model.predict(X_test)

# 3. Print the first 10 predicted values and compare them with actual values
# We convert actual values to a list to easily print side-by-side
actual_10 = list(y_test.head(10))
predicted_10 = list(y_pred[:10])

print("\n--- Q6: Comparing First 10 Predictions ---")
print(f"{'Actual Price':<15} | {'Predicted Price':<15}")
print("-" * 35)
for actual, pred in zip(actual_10, predicted_10):
    print(f"${actual:<14,.2f} | ${pred:<14,.2f}")
```

```python
# --- Q7. Model Evaluation using R² Score ---

# 1. Calculate the R² score using r2_score(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n--- Q7: Model Evaluation ---")
print(f"R² Score: {r2:.4f}")

# 2 & 3. Interpret the R² value and comment on performance
# Note: In your final assignment submission, rewrite these text comments in your own words!
print("\nInterpretation & Performance Comment:")
print(
    f"An R² score of {r2:.2f} means that approximately {r2*100:.1f}% of the variation "
)
print("in car prices can be successfully explained by our model's features.")
if r2 > 0.8:
    print(
        "Performance Comment: The model is performing very well with high predictive accuracy."
    )
elif r2 > 0.6:
    print(
        "Performance Comment: The model is performing moderately well, but has room for improvement."
    )
else:
    print(
        "Performance Comment: The model is performing poorly and needs better features or a different algorithm."
    )
```

```python
# --- Q8. Saving the Model ---

# Save your trained model and preprocessing objects using joblib
# 1. Save Linear Regression model -> "LR_ford_car.pkl"
joblib.dump(model, "LR_ford_car.pkl")

# 2. Save StandardScaler -> "scaler.pkl"
joblib.dump(scaler, "scaler.pkl")

# 3. Save the list of columns -> "columns.pkl"
# Saving the exact column order ensures future raw data is encoded perfectly
joblib.dump(list(X_train.columns), "columns.pkl")

print("\n--- Q8: Files Saved ---")
print("Successfully saved: LR_ford_car.pkl, scaler.pkl, and columns.pkl")
```

```python
import joblib
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ==========================================
# 1. LOAD THE DATASET
# ==========================================
# Change 'your_other_car_dataset.csv' to your actual file name
df = pd.read_csv("/content/cardata.csv")

# Identify your target column name (e.g., 'price' or 'selling_price')
# Replace 'price' below if your target column has a different name
target_column = "Selling_Price"

# ==========================================
# 2. SEPARATE FEATURES AND TARGET
# ==========================================
X = df.drop(columns=[target_column])
Y = df[target_column]

# ==========================================
# 3. ONE-HOT ENCODING (Categorical to Numbers)
# ==========================================
# Automatically find text columns (like brand, fuel type, transmission)
categorical_cols = X.select_dtypes(include=["object", "category"]).columns

# Convert text categories into 1s and 0s
X_encoded = pd.get_dummies(X, columns=categorical_cols)
X_encoded = X_encoded.astype(int)

# ==========================================
# 4. FEATURE SCALING
# ==========================================
# Automatically find all numerical features left in the dataset
numerical_cols = X_encoded.select_dtypes(include=["int64", "float64"]).columns

X_scaled = X_encoded.copy()
scaler = StandardScaler()

# Scale numerical columns to normalize the ranges
X_scaled[numerical_cols] = scaler.fit_transform(X_scaled[numerical_cols])

# ==========================================
# 5. TRAIN-TEST SPLIT
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, Y, test_size=0.33, random_state=42
)

# ==========================================
# 6. MODEL TRAINING
# ==========================================
model = LinearRegression()
model.fit(X_train, y_train)

# ==========================================
# 7. EVALUATION
# ==========================================
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print("--- Optional Task Model Summary ---")
print(f"Dataset Size: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"R² Score on Test Data: {r2:.4f}")

# ==========================================
# 8. SAVE THE NEW ML PIPELINE
# ==========================================
# Saving with unique names so it doesn't overwrite your Ford model files
joblib.dump(model, "LR_optional_car_model.pkl")
joblib.dump(scaler, "optional_scaler.pkl")
joblib.dump(list(X_train.columns), "optional_columns.pkl")

print("\nSaved new pipeline files successfully!")
```
